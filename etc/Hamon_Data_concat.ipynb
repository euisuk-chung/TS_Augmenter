```python
# data_path = '/repo/projects/HAI_timeseries/Hamon_dataset/*.csv'
data_path = '../data/*.csv'
```

```python
file_list = glob.glob(data_path)
file_list.sort()
file_list
```

```python
# 전체 데이터
df_total = pd.DataFrame()

for i in file_list:
    data = pd.read_csv(i)
    df_total = pd.concat([df_total, data])

df_total = df_total.reset_index(drop = True)
```
